```python
json_dir = "/content/dataset"

from operator import itemgetter        
import csv
import json
import os

header = ['artist', 'timestamp', 'similars', 'tags', 'track_id', 'title']
required_cols = itemgetter(*header)
count = 0
# loop through each directory and file in the JSON directory tree
for root, dirs, files in os.walk(json_dir):
    for filename in files:
        if filename.endswith('.json'):
            print(os.path.join(root, filename))
            try:
                with open(os.path.join(root, filename)) as f_input, open('/content/lastfm_subset_dataset1.csv', 'a', newline='') as f_output:
                    csv_output = csv.writer(f_output)
                    if count == 0:
                        csv_output.writerow(header)
                        count += 1
                    for row in f_input:
                        if row.strip():
                            csv_output.writerow(required_cols(json.loads(row)))
            except Exception as e:
                print(f"Error processing file '{filename}': {e}")
```
